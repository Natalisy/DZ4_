```python
documents = [
{'type':'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
{'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
{'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

directories = {
'1': ['2207 876234', '11-2'],
'2': ['10006'],
'3': []
}


def launch():
	print("Введите команду: ")
	whats_up_secretary = input()

	if whats_up_secretary == "p":
		secretary_help()

	elif whats_up_secretary == "s":
		shelf_cheaking()

	elif whats_up_secretary == "i":
		cheaking_info()

	elif whats_up_secretary == "as":
		add_shelf()

	elif whats_up_secretary == "ds":
		delete_shelf()

	elif whats_up_secretary == "q":
		return whats_up_secretary


def whats_doc():
	print("Введите номер документа: ")
	document_number = input()
	#doc_cheaking(document_number)
	return document_number


"""def doc_cheaking(document_number):
	if document_number not in documents:
		return "Документ не найден в базе"
	else:
		return None"""


def secretary_help():
	document_number = whats_doc()
	flag = False
	for doc in documents:
		if doc['number'] == document_number:
			print("Владелец документа:", doc['name'])
			flag = True
	if flag == False:
		print("Документ не найден в базе")


def shelf_cheaking():
	document_number = whats_doc()
	flag = False
	for shelf in directories:
		if document_number in directories[shelf]:
			print("Документ хранится на полке:", shelf)
			flag = True
	if flag == False:
		print("Документ не найден в базе")


def cheaking_info():
	for direct in directories:
		for doc_list in directories[direct]:
			information = ''
			information += f'№:{doc_list}, '
			for doc in documents:
				if doc['number'] == doc_list:
					information += f'тип: {doc["type"]}, владелец: {doc["name"]}, полка хранения: {direct}'
					print(information)


def add_shelf():
	print("Введите номер полки: ")
	shelf_number = input()
	shelfs = ''

	if shelf_number not in directories:
		directories[shelf_number] = []
		for i in directories:
			shelfs += i
		print("Полка добавлена. Текущий перечень полок: " + ','.join(shelfs) + '.')

	else:
		for i in directories:
			shelfs += i
		print("Такая полка уже существует. Текущий перечень полок: " + ','.join(shelfs) + '.')


def delete_shelf():
	print("Введите номер полки: ")
	shelf_number = input()
	shelfs = ''

	if shelf_number not in directories:
		for i in directories:
			shelfs += i
		print("Такой полки не существует. Текущий перечень полок: "  + ','.join(shelfs) + '.')
		return

	if directories[shelf_number] == []:
		del directories[shelf_number]
		for i in directories:
			shelfs += i
		print("Полка удалена. Текущий перечень полок: " + ','.join(shelfs) + '.')
	elif directories[shelf_number] != []:
		for i in directories:
			shelfs += i
		print("На полке есть документы, удалите их перед удалением полки. Текущий перечень полок: " + ','.join(shelfs) + '.')


while True:
	launch()
	if launch() == "q":
		break
        
        
```
